DRAWS A BOX

```python
import fitz

def draw_rectangle_on_pdf(pdf_path, page_number, coordinates, output_pdf_path):
    document = fitz.open(pdf_path)
    page = document.load_page(page_number)
    
    # Draw a rectangle
    rect = fitz.Rect(coordinates)
    page.draw_rect(rect, color=(1, 0, 0), width=1.5)  # Red rectangle

    # Save to a new PDF
    document.save(output_pdf_path)
    document.close()

pdf_path = '1update_NonMD95.pdf'
coordinates = (72, 340, 84, 420)  # Example coordinates (x1, y1, x2, y2)
page_number = 0  # Page number where you want to draw the rectangle

# Format the output file name to include the coordinates
output_pdf_path = f'rectangle_{coordinates[0]}_{coordinates[1]}_{coordinates[2]}_{coordinates[3]}.pdf'

draw_rectangle_on_pdf(pdf_path, page_number, coordinates, output_pdf_path)
```

DRAWS A GRIDLINE

```python
import fitz

def draw_grid_on_pdf(pdf_path, page_number, grid_spacing, output_pdf_path):
    document = fitz.open(pdf_path)
    page = document.load_page(page_number)

    # Get page dimensions
    page_width = page.rect.width
    page_height = page.rect.height

    # Draw horizontal and vertical gridlines
    for x in range(0, int(page_width), grid_spacing):
        start_point = fitz.Point(x, 0)
        end_point = fitz.Point(x, page_height)
        page.draw_line(start_point, end_point, color=(0, 0, 1), width=0.5)

        # Optionally add coordinate labels
        page.insert_text((x, 10), f'x={x}', color=(1, 0, 0), fontsize=8)

    for y in range(0, int(page_height), grid_spacing):
        start_point = fitz.Point(0, y)
        end_point = fitz.Point(page_width, y)
        page.draw_line(start_point, end_point, color=(0, 0, 1), width=0.5)

        # Optionally add coordinate labels
        page.insert_text((10, y), f'y={y}', color=(1, 0, 0), fontsize=8)

    # Save to a new PDF
    document.save(output_pdf_path)
    document.close()

pdf_path = '1Q3Non-MD_95% 22.pdf'
output_pdf_path = 'output_with_grid.pdf'
grid_spacing = 50  # Grid spacing in points
page_number = 0  # Page number to draw the grid on

draw_grid_on_pdf(pdf_path, page_number, grid_spacing, output_pdf_path)
```

```python
import fitz
import re
import os
import shutil

def extract_text_from_coordinates(pdf_path, coordinates):
    document = fitz.open(pdf_path)
    extracted_text = []

    for page_number in range(len(document)):
        page = document.load_page(page_number)
        text = page.get_text("text", clip=coordinates)
        extracted_text.append(text.strip())

    document.close()
    return extracted_text

def sanitize_filename(text):
    text = re.sub(r'[\\/*?:"<>|]', "", text)
    text = text.replace('\n', '_').replace('\r', '_').strip()
    return text[:50]

def split_pdf_page(source_pdf_path, page_number, output_pdf_name):
    pdf_reader = fitz.open(source_pdf_path)
    pdf_writer = fitz.open()

    # Rotate the page 90 degrees clockwise
    page = pdf_reader.load_page(page_number)
    page.set_rotation(90)

    pdf_writer.insert_pdf(pdf_reader, from_page=page_number, to_page=page_number)
    pdf_writer.save(output_pdf_name)
    pdf_writer.close()

# Path to your PDF and the directory to store files
pdf_path = '1updateMD.pdf'
directory_path = 'C:/Users/cwchen/Documents/Time Study/Updated/Updated Time_Sheet MD'

name_coordinates = (60, 630, 75, 900)

# Extract and split PDF
names = extract_text_from_coordinates(pdf_path, name_coordinates)
for page_number, name in enumerate(names):
    sanitized_name = sanitize_filename(name)
    output_pdf_name = f"{directory_path}/{sanitized_name}.pdf"
    split_pdf_page(pdf_path, page_number, output_pdf_name)
    print(f"Created PDF: {output_pdf_name}")

# Organize files into alphabetical folders
for filename in os.listdir(directory_path):
    if os.path.isfile(os.path.join(directory_path, filename)):
        first_letter = filename[0].upper()
        if first_letter.isalpha():
            folder_path = os.path.join(directory_path, first_letter)
            if not os.path.exists(folder_path):
                os.makedirs(folder_path)
            shutil.move(os.path.join(directory_path, filename), folder_path)

print("Files have been organized into alphabetical folders.")
```

```python
import fitz  # PyMuPDF
import re
import csv

def extract_text_from_coordinates(pdf_path, coordinates):
    document = fitz.open(pdf_path)
    extracted_text = []

    for page_number in range(len(document)):
        page = document.load_page(page_number)
        text = page.get_text("text", clip=coordinates)
        extracted_text.append(text.strip())

    document.close()
    return extracted_text

def sanitize_filename(text):
    text = re.sub(r'[\\/*?:"<>|]', "", text)
    text = text.replace('\n', '_').replace('\r', '_').strip()
    return text[:50]

# Path to your PDF
pdf_path = '1update_NonMD95.pdf'

# Coordinates to extract the text (modify as needed)
text_coordinates = (72, 340, 84, 420)

# Extract text
extracted_text = extract_text_from_coordinates(pdf_path, text_coordinates)

# Sanitize and prepare the text for CSV
sanitized_text = [sanitize_filename(text) for text in extracted_text]

# Path for the output CSV file
csv_file_path = 'extract_nonmd_update'

# Write the extracted text to a CSV file
with open(csv_file_path, mode='w', newline='', encoding='utf-8') as file:
    writer = csv.writer(file)
    for row in sanitized_text:
        writer.writerow([row])

print("Extracted text has been saved to CSV.")

```
