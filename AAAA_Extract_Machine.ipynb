```python
import PyPDF2
import re

def extract_information(text):
    name_match = re.search(r"\d{2}/\d{2}/\d{4}\s*\n([A-Z\s,]+)\s*Activity", text)
    id_match = re.search(r"Job Classification/Position\s*\n.*?(\d{7,9})", text)

    if name_match and id_match:
        name = name_match.group(1).strip().replace(",", "").replace(" ", "_")
        employee_id = id_match.group(1).strip()
        return f"{name}_{employee_id}"
    else:
        return None


with open('Timesheet by UCLA 2024Q2 NonMD - NONPHY - 628.pdf', 'rb') as file:
    reader = PyPDF2.PdfReader(file)
    missing_info_pages = []

    for i, page in enumerate(reader.pages):
        text = page.extract_text()
        filename = extract_information(text) if text else None

        if filename is None:
            missing_info_pages.append(i + 1)
            filename = f"Page_{i + 1}_Information_Not_Found"

        writer = PyPDF2.PdfWriter()
        writer.add_page(page)

        with open(f"{filename}.pdf", 'wb') as output_pdf:
            writer.write(output_pdf)

# Optionally, print or save the list of pages with missing information
print("Pages with missing information:", missing_info_pages)
# You can also write this list to a file if needed
```
